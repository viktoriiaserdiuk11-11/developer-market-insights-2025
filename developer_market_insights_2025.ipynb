```python
import pandas as pd
# define programming languages for comparison
languages = [
    'Python',
    'JavaScript',
    'SQL',
    'HTML/CSS',
    'TypeScript',
    'C++',
    'C#',
    'PHP',
    'Go',
    'Rust'
]

language_counts = []
# count respondents for each language
for lang in languages:
    count = df['LanguageHaveWorkedWith'].str.contains(
        lang,
        case=False,
        na=False
    ).sum()

    language_counts.append([lang, count])
# create and sort language popularity table
languages_top10 = pd.DataFrame(
    language_counts,
    columns=['Language', 'Count']
)

languages_top10 = languages_top10.sort_values(
    by='Count',
    ascending=False
).reset_index(drop=True)

languages_top10
```

```python
# export top 10 programming languages to CSV
languages_top10.to_csv('languages_top10.csv', index=False)
```

```python
# calculate median salary and number of respondents by country

country_stats = py_df.groupby('Country')['ConvertedCompYearly'].agg(
    median_salary='median',
    respondents='count'
)

# filter countries with at least 30 respondents

country_stats = country_stats[
    country_stats['respondents'] >= 30
]

# sort countries by median salary in descending order
# select top 10 countries

country_stats = country_stats.sort_values(
    by='median_salary',
    ascending=False
).head(10)

country_stats
```

```python
# convert country statistics index to columns
country_stats = country_stats.reset_index()

# export country salary statistics to CSV
country_stats.to_csv(
    'salary_by_country_filtered_top10.csv',
    index=False
)

country_stats
```

```python
# convert work format summary to table format
remote_chart = remote_summary_clean.reset_index()

# rename columns for dashboard use
remote_chart.columns = ['Work_Format', 'Respondents']

# simplify work format labels
remote_chart['Work_Format'] = remote_chart['Work_Format'].replace({
    'Hybrid (some remote, leans heavy to in-person)': 'Hybrid Remote',
    'Hybrid (some in-person, leans heavy to flexibility)': 'Hybrid Flexible',
    'Your choice (very flexible, you can come in when you want or just as needed)': 'Flexible Choice'
})

print(remote_chart)
```
